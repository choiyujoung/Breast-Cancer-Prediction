# 1. Import Library and load data
데이터 분석과 머신러닝 모델 구축에 필요한 핵심 라이브러리들을 불러옵니다.

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

cancer = load_breast_cancer()
```

# 2. 데이터 전처리 (X와 y 설정)
머신러닝에서 데이터를 나눌 때의 핵심은 '무엇으로(X) 무엇을(y) 예측할 것인가'입니다.

X (Features): 예측을 위해 모델에 입력하는 데이터입니다. 보통 2차원 배열(행렬) 형태이며 여러 개의 컬럼을 가집니다.
y (Target/Label): 모델이 맞히고 싶은 정답이며, 보통 1차원 배열(벡터) 형태입니다. 
 
여기서는 데이터 조작의 편의성을 위해 Pandas DataFrame을 사용합니다. DataFrame은 엑셀 시트와 비슷하여, 전체 표를 데이터프레임으로, 각 열(Column)을 특징(Features)으로 이해할 수 있습니다. 

DataFrame을 사용하여 특징(Features)을 X로, 악성/양성 여부를 y로 설정했습니다.

```python
X = pd.DataFrame(cancer.data,columns=cancer.feature_names)

y = cancer.target
```

# 3. 데이터 분할 및 모델 학습
전체 데이터를 모델 학습용(Train)과 성능 검증용(Test)으로 분리합니다

```python
#divide data into test and train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)


model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

# 4. 모델 성능 평가
학습된 모델이 새로운 데이터(X_test)에 대해 얼마나 잘 예측하는지 정확도를 확인합니다

```python
#predictions
prediction = model.predict(X_test)

score = accuracy_score(y_test, prediction)
print(f"모델 정확도: {score}") 
```

# 5. 모델 시각화 (Decision Tree structure)
tree.export_graphviz 함수는 모델이 어떤 기준으로 데이터를 분류했는지 보여주는 설계도를 만듭니다. 


feature_names (질문 리스트): X를 DataFrame으로 만들었기 때문에 X.columns에서 확인 가능합니다. 


class_names (결과 카테고리): 분류 결과가 '악성/양성' 문자로 표시되도록 cancer.target_names를 지정합니다. 

최근에는 더 직관적인 tree.plot_tree가 널리 사용됩니다.

```python
from sklearn import tree
#이 부분에서 feature_names and class_names를 어떻게 설정해야할지 모르겠었다.
tree.export_graphviz(model,
                     out_file='breast-cancer.dot',
                     feature_names=list(X.columns),
                     class_names= cancer.target_names,
                     label='all',
                     rounded=True,
                     filled=True)
```

최근에는 더 직관적인 tree.plot_tree가 널리 사용됩니다.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
tree.plot_tree(model,
							feature_names=list(X.columns),
							class_names=list(cancer.target_names),
							filled=True)
plt.show()
```

# 6. 모델 저장
학습된 모델을 파일로 저장하여 언제든 다시 사용할 수 있도록 합니다.

```python
import joblib
model = joblib.dump(model, 'breast-cancer.joblib')
```
